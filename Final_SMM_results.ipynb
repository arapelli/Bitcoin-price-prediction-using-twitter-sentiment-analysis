```python
def positive(x):
    if x[0] + x[1] + x[2] >=1:
        y=1
    
    else:
        y=0
    
    return y
        
```

```python
dft["3_pe"] = dft["3_p"].apply(positive)
```

```python
dft["3_n"] = dft[["A_N","TB_N","V_N"]].values.tolist()
```

```python
def negative(x):
    if x[0] + x[1] + x[2] <= -1:
        y=-1
    
    else:
        y=0
    
    return y
```

```python
dft["3_ne"] = dft["3_n"].apply(negative)
```

```python
def sent(x):
    y=0
    if x[1] == 1:
        y= -1
    elif x[0] == 1:
        y=1
    elif x[2] + x[3] == 0:
        y=0
    elif x[2] + x[3] < 0:
        y=-1
    else:
        y=1
    
    return y
        
```
